```python
# check if there are null values in the dataframe
null_columns=df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
```

```python
# dropped null values
df = df.dropna(how='any',axis=0) 
```

```python
df.info()
```

```python
df.to_csv('cleaned_craigslistVehiclesFull.csv')
```

```python
#drops unwated columns

cols = ['vin','county_fips','county_name','state_fips','state_code','state_name','weather','image_url','lat','long','url']

dropped_df = df.drop(columns = cols)
```

```python
null_columns=dropped_df.columns[dropped_df.isnull().any()]
dropped_df[null_columns].isnull().sum()
```

```python
# reset dataframe index and drop the original index column

dropped_df.reset_index(inplace = True)
dropped_df = dropped_df.drop("index", axis=1)
#dropped_df = dropped_df.drop("level_0", axis=1)

dropped_df.head()
```

```python
dropped_df.dtypes
```

```python
# change column year from float to int
dropped_df.year = dropped_df.year.astype(int)
dropped_df.odometer = dropped_df.odometer.astype(int)
```

```python
dropped_df.head()
```

```python
dropped_df[:1]
```

```python
# the 'make' column is really messay
dropped_df['make'].unique()
```

```python
# check for column values
dropped_df['manufacturer'].unique()
```

```python
# clean the manufacturer column values - brands were referred as differnt names
dropped_df['manufacturer'].replace(['chevy','chev'], 'chevrolet',inplace=True)
dropped_df['manufacturer'].replace(['mercedes-benz'], 'mercedes',inplace=True)
dropped_df['manufacturer'].replace(['alfa'], 'alfa-romeo',inplace=True)
dropped_df['manufacturer'].replace(['harley'], 'harley-davidson',inplace=True)
dropped_df['manufacturer'].replace(['landrover'], 'land rover',inplace=True)
dropped_df['manufacturer'].replace(['infinity'], 'infiniti',inplace=True)
dropped_df['manufacturer'].replace(['aston'], 'aston-martin',inplace=True)
dropped_df['manufacturer'].replace(['vw'], 'volkswagen',inplace=True)
```
